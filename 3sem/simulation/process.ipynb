```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import subprocess
import json
```

```python
config_path = "config.json"
out_path = "out.csv"
config = {
    "dt": 0.05, 
    "x_0": 1, 
    "v_0": 0, 
    "w_0": 1, 
    "gamma": 5, 
    "w": 0.5, 
    "f_0": 1, 
    "n_steps": 1000,
    "solver": "RK4",
    "equation": "sq1"
}

with open(config_path, 'w') as f:
    json.dump(config, f)

subprocess.run(['build/main', config_path, out_path])
df = pd.read_csv(out_path)
```

```python
plt.plot(df.time, df.position)
plt.xlabel('time')
plt.ylabel('pos')
```
