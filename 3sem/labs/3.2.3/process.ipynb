```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## АЧХ

```python
df1 = pd.read_excel('table1.xlsx')
df1.head()
```

```python
df2 = pd.read_excel('table2.xlsx')
df2.head()
```

```python
plt.rc('font', size=22)

markers = ['o', 'v', '>', '<', '^', 's', 'D']
# markers = ['1', '2', '3', '4', '+', 'x', '.']

fig, ax = plt.subplots(figsize=(14, 10))
ax.spines['bottom'].set_linewidth(2)
ax.spines['top'].set_linewidth(2)
ax.spines['right'].set_linewidth(2)
ax.spines['left'].set_linewidth(2)
ax.grid()

ax.plot(df1.f, df1.U, linewidth=2, c='black', marker='o', label='1')
ax.plot(df2.f, df2.U, linewidth=2, c='black', marker='v', label='2')

ax.errorbar(df1.f, df1.U, xerr = df1.ef, yerr = df1.eU, fmt='none', linewidth=2, c='black')
ax.errorbar(df2.f, df2.U, xerr = df2.ef, yerr = df2.eU, fmt='none', linewidth=2, c='black')
ax.legend()

ax.set_ylabel(r"$U, В$")
ax.set_xlabel(r"$f, кГц$")
fig.savefig(r'images/Uf.eps', format='eps')
```

```python
plt.rc('font', size=22)

markers = ['o', 'v', '>', '<', '^', 's', 'D']
# markers = ['1', '2', '3', '4', '+', 'x', '.']

fig, ax = plt.subplots(figsize=(14, 10))
ax.spines['bottom'].set_linewidth(2)
ax.spines['top'].set_linewidth(2)
ax.spines['right'].set_linewidth(2)
ax.spines['left'].set_linewidth(2)
ax.grid()

ax.plot(df1.f / 23.21, df1.U / 0.8235, linewidth=2, c='black', marker='o', label='1')
ax.plot(df2.f / 19.47, df2.U / 0.5655, linewidth=2, c='black', marker='v', label='2')

ax.errorbar(df1.f / 23.21, df1.U / 0.8235, xerr = df1.ef / 23.21, yerr = 2**0.5 * df1.eU / 0.8235, fmt='none', linewidth=2, c='black')
ax.errorbar(df2.f / 19.47, df2.U / 0.5652, xerr = df2.ef / 19.47, yerr = 2**0.5 * df2.eU / 0.5655 , fmt='none', linewidth=2, c='black')
ax.legend()

ax.set_ylabel(r"$U / U_0$")
ax.set_xlabel(r"$f / f_0$")
fig.savefig(r'images/Uf_norm.eps', format='eps')
```

### ФЧХ

```python
a = np.pi / 4 #константа подгона
```

```python
df1 = pd.read_excel('table3.xlsx')
df1.P[:8] -= np.pi *a
df1.head()
```

```python
df2 = pd.read_excel('table4.xlsx')
df2.P[:10] -= np.pi *a
df2.head()
```

```python
plt.rc('font', size=22)

markers = ['o', 'v', '>', '<', '^', 's', 'D']
# markers = ['1', '2', '3', '4', '+', 'x', '.']

fig, ax = plt.subplots(figsize=(14, 10))
ax.spines['bottom'].set_linewidth(2)
ax.spines['top'].set_linewidth(2)
ax.spines['right'].set_linewidth(2)
ax.spines['left'].set_linewidth(2)
ax.grid()

ax.plot(df1.f / 23.21, df1.P / np.pi, linewidth=2, c='black', marker='o', label='1')
ax.plot(df2.f / 19.47, df2.P / np.pi, linewidth=2, c='black', marker='v', label='2')

ax.errorbar(df1.f / 23.21, df1.P / np.pi, xerr = df1.ef / 23.21, yerr = df1.eP / np.pi, fmt='none', linewidth=2, c='black')
ax.errorbar(df2.f / 19.47, df2.P / np.pi, xerr = df2.ef / 19.47, yerr = df2.eP / np.pi , fmt='none', linewidth=2, c='black')
ax.legend()

ax.set_ylabel(r"$\varphi / \varphi_0$")
ax.set_xlabel(r"$f / f_0$")
fig.savefig(r'images/pf_norm.eps', format='eps')
```
