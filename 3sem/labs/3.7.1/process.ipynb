```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df1 = pd.read_excel('1.xlsx')
df1.head()
```

```python
from scipy.optimize import curve_fit

function = lambda x, a, b: a + (b**2)*x
popt, pcov = curve_fit(function, xdata = df1.nu2, ydata = df1['1D2'], sigma = df1.e1D2)
popt, np.sqrt(pcov)
```

```python
plt.rc('font', size=22)

markers = ['o', 'v', '>', '<', '^', 's', 'D']
# markers = ['1', '2', '3', '4', '+', 'x', '.']

fig, ax = plt.subplots(figsize=(14, 10))
ax.spines['bottom'].set_linewidth(2)
ax.spines['top'].set_linewidth(2)
ax.spines['right'].set_linewidth(2)
ax.spines['left'].set_linewidth(2)
ax.grid()

ax.scatter(df1.nu2, df1['1D2'], linewidth=1, c='black', marker='o')
ax.errorbar(df1.nu2, df1['1D2'], xerr = df1.enu2, yerr = df1.e1D2, fmt='none', linewidth=2, c='black')
plt.plot(np.array([0, 205000]), function(np.array([0, 205000]), popt[0], popt[1]), c='black')
# ax.legend()

ax.set_ylabel(r"$1/\zeta^2, Гц^2/Ом^2$")
ax.set_xlabel(r"$\nu^2, Гц^2$")
fig.savefig(r'zetanu.png', format='png')
```
