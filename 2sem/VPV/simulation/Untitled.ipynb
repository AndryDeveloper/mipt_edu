```python
from numpy import genfromtxt
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.rc('font', size=22)

markers = ['o', 'v', '>', '<', '^', 's', 'D']
# markers = ['1', '2', '3', '4', '+', 'x', '.']

fig, ax = plt.subplots(figsize=(14, 10))
ax.spines['bottom'].set_linewidth(2)
ax.spines['top'].set_linewidth(2)
ax.spines['right'].set_linewidth(2)
ax.spines['left'].set_linewidth(2)
ax.grid()

k = 0
ks_ = []
for i in range(5, 13, 2):
    data = genfromtxt(f'R{i}D0N300SV2', delimiter=',')
    a, b = np.linalg.lstsq(np.vstack([data[:, 0], np.ones(len(data[:, 0]))]).T, data[:, 1], rcond=None)[0]
    ks_.append(a)
    ax.plot(data[:, 0], data[:, 1], linewidth=2, c='black', marker=markers[k], label=str(k + 1))

    X = np.array([min(data[:, 0])*0, max(data[:, 0])*1.03])
    Y = b + a*X
    ax.plot(X, Y, linewidth=2, c='gray', dashes=[0, 0, 2])
    
    k += 1

ax.legend()

ax.set_ylabel(r"$F$")
ax.set_xlabel(r"$N$")
fig.savefig(r'../figures/FN.eps', format='eps')
```

```python
bs_ = [5, 7, 9, 11]
ks = ks_
bs = np.array(bs_)**2

plt.rc('font', size=22)

markers = ['o', 'v', '>', '<', '^', 's', 'D']
# markers = ['1', '2', '3', '4', '+', 'x', '.']

fig, ax = plt.subplots(figsize=(14, 10))
ax.spines['bottom'].set_linewidth(2)
ax.spines['top'].set_linewidth(2)
ax.spines['right'].set_linewidth(2)
ax.spines['left'].set_linewidth(2)
ax.grid()

a, b = np.linalg.lstsq(np.vstack([bs, np.ones(len(bs))]).T, ks, rcond=None)[0]
ax.scatter(bs, ks, linewidth=2, c='black', marker='o')

X = np.array([min(bs)*0.97, max(bs)*1.03])
Y = b + a*X
ax.plot(X, Y, linewidth=2, c='gray', dashes=[0, 0, 2])


ax.set_ylabel(r"$k$")
ax.set_xlabel(r"$d^2$")
fig.savefig(r'../figures/kr.eps', format='eps')
```
