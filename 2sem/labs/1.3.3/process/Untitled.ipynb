```python
from gradient_errors import Environment
import matplotlib.pyplot as plt
import numpy as np
import torch
import pandas as pd
```

```python
env = Environment('data_normal.xlsx')
```

```python
lnQ, lnP = torch.log(env.T11.Q), torch.log(env.T11.P)
a1, b1 = env.chi2(lnP[:8], lnQ[:8])
a2, b2 = env.chi2(lnP[-5:], lnQ[-5:])
a3, b3 = env.chi2(env.T11.P[:8], env.T11.Q[:8], mode='zero')
```

```python
nu1 = np.pi * (env.d1/2)**4 / (8 * env.l1 * b3 / 60 / 1000)
```

```python
plt.rc('font', size=22)

markers = ['o', 'v', '>', '<', '^', 's', 'D']
# markers = ['1', '2', '3', '4', '+', 'x', '.']

fig, ax = plt.subplots(figsize=(14, 10))
ax.spines['bottom'].set_linewidth(2)
ax.spines['top'].set_linewidth(2)
ax.spines['right'].set_linewidth(2)
ax.spines['left'].set_linewidth(2)
ax.grid()

Q, P = env.T11.Q.detach().numpy(), env.T11.P.detach().numpy()
ax.scatter(np.log(P)[:8], np.log(Q)[:8], linewidth=2, c='black', marker='o', label='1')
ax.scatter(np.log(P)[-5:], np.log(Q)[-5:], linewidth=2, c='black', marker='v', label='2')
ax.scatter(np.log(P)[8], np.log(Q)[8], linewidth=2, c='black', marker='>', label='3')

X = np.array([min(np.log(P)[:8])*0.97, max(np.log(P)[:8])*1.03], )
Y = a1.item() + b1.item()*X
ax.plot(X, Y, linewidth=2, c='gray', dashes=[0, 0, 2])

X = np.array([min(np.log(P)[-5:])*0.97, max(np.log(P)[-5:])*1.03])
Y = a2.item() + b2.item()*X
ax.plot(X, Y, linewidth=2, c='gray', dashes=[0, 0, 2])

ax.errorbar(np.log(P), np.log(Q), xerr = env.get_error(lnP), yerr = env.get_error(lnQ), fmt='none', linewidth=2, c='black')

ax.legend()

ax.set_ylabel(r"$ln Q$")
ax.set_xlabel(r"$ln P$")
fig.savefig(r'../images/lnPlnQd1.eps', format='eps')
```
